# Sets

- Sets are unordered collection of objects (no dublicates)
- Each object object is called element
-  



Examples: 
$$
3 \in \mathbb{Z}, \pi \in \mathbb{R}, \mathbb{Z} \subseteq \mathbb{R}
$$

```
def myfunc(input_list):
	output = [element for element in input_list if element%2 != 0 if element%3 != 0 if element%5 != 0]  
	print(output)

myfunc(range(1,8))
```

### #returns elements not divisible by 2,3,5

```
def myfunc(input_list):
	output = {element for element in input_list if (element%2 != 0) and (element%3 != 0) and (element%5 != 0)}  #notice curly braces
	print("out")
	print(output)
```

```
myfunc(range(1,8))


#Example of multiple ifs: 
#[i for i in range(100) if i > 10 if i < 50]
```

```

#--------------------------------------------------------

import re

def myfunc2(strs):
	output = {x for x in strs if re.match(r'^[a-mA-M]*S',x)}
	print(output)

print(re.match(r"^[a-mA-M]*S","dig"))

myfunc2({'cat','dig','dog','ant','eel'})
```
